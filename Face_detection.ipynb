```
 import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mimg
samp=6
train_data=np.zeros((5*6,144*120*3))
train_label=np.zeros((6*5))
test_data=np.zeros((5,144*120*3))
test_label=np.zeros((5))
count=-1
plt.figure(1)
plt.ion()
for i in range(1,6): 
  for j in range(1,samp+1):
   plt.cla()
   count+=1
   path='/content/drive/My Drive/d/f%d/%d.png'%(i,j)
   im=mimg.imread(path)
   feat=im.reshape(1,-1)
   train_data[count,:]=feat
   train_label[count]=i
   plt.imshow(im,cmap='gray')
   plt.title('user no:'+str(i)+'sample no:'+str(j))
   #plt.pause(0.1)
count=-1
for i in range(1,6): 
   plt.cla()
   count+=1
   path='/content/drive/My Drive/d/f%d/%d.png'%(i,7)
   im=mimg.imread(path)
   feat=im.reshape(1,-1)
   test_data[count,:]=feat
   test_label[count]=i
   plt.imshow(im,cmap='gray')
   plt.title('test sample:%d'%(i))
   plt.pause(0.1)
```

```
incorrect=0;
correct=0;
for i in range(0,5):
    n=int(input('enter test sample no(1-5) to identify & get name:'))
    test_sample=test_data[n-1,:]
    actual_label=test_label[n-1]
    #print(test_sample.shape)
   
    #print(actual_label)
    d = np.zeros((samp*5))
    #print(d)
    for j in range(0,samp*5):
        
        d[j] = np.sum((test_sample-train_data[j,:])**2)
        #print(d[j])
    # take index of minium distance
    val = d.min()
    ind = d.argmin()
    pred_op=train_label[ind]
    if pred_op==actual_label:
        correct=correct+1 
        if(n==1):
          print('correctly matched,Name:Dhulqar salman')
        elif(n==2):
          print('correctly matched,Name:Surya')
        elif(n==3):
          print('correctly matched,Name:Sonam kapoor') 
        elif(n==4):
          print('correctly matched,Name:Ranbeer kapoor')      
        elif(n==5):         
          print('correctly matched,Name:Prabhas')     
    else:
        incorrect = incorrect+1
        print('wrongly matched')
accuracy = (correct/(incorrect+correct))*100
print('Accruacy of face recognition with 70:30 ratio using distance formula is ',accuracy)
```
